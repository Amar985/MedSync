# Medsync Data Exploration

```python
import pandas as pd
DATA_PATH = "../data"
```

```python
df = pd.read_excel(DATA_PATH + "/consumo_material_clean.xlsx")
print(df.shape)
```

```python
df.head(10)
```

```python
df['H_code'] = df['H_code'].str.replace('--', '-')
df[['REGION', 'HOSPITAL', 'DEPARTMENT0']] = df['H_code'].str.split('-', expand=True)
df.drop(columns=['H_code'], inplace=True)
```

```python
# lookup all products with P_date = 01/01/23
# sort df by P_date (year with two digits)

df['P_date'] = pd.to_datetime(df['P_date'], format='%d/%m/%y')
df.sort_values(by='P_date', inplace=True)
df.reset_index(drop=True, inplace=True)

df.head(10)
```

```python
cols = ['P_code', 'P_date', 'N_Order', 'R_no', 'N_product_purchased',
       'N_units', 'Cost', 'Total_product_purchase', 'Type_purchase',
       'H_code', 'Type_logistic', 'PRODUCTO']

# check if P_code and R_no are a bijection
def check_bijection(col1, col2):
    if col1 not in df.columns or col2 not in df.columns:
        return False
    g1 = df.groupby(col1)[col2].nunique().value_counts()
    g2 = df.groupby(col2)[col1].nunique().value_counts()
    return len(g1) == 1 and len(g2) == 1 and g1.index[0] == 1 and g2.index[0] == 1

# for each pair of columns in cols, check if they are a bijection and print the column names
for i in range(len(cols)):
    for j in range(i+1, len(cols)):
        if check_bijection(cols[i], cols[j]):
            print(cols[i], cols[j])

# show the R_no values where there is a character
print(df[df['R_no'].str.contains('[a-zA-Z]', na=False)]['R_no'].unique())
```

```python
# check if the prices by PRODUCTO are the same
df['AÑO'] = df['P_date'].dt.year
g = df.groupby(['PRODUCTO'])['N_units'].unique()
print(g)
```
